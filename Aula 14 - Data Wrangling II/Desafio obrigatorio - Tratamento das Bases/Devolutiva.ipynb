```python
# Outro ponto seria a parte de salvar as tabelas transformadas em um banco de dados local utilizando sqlite3. Para isso, você pode utilizar aquelas funções que criamos na aula de banco de dados, em que você passa o dataframe na função e ela armazenará este dataframe no banco de dados. Assim, você conseguirá salvar suas tabelas tratadas no banco de dados sqlite.  As funções de banco de dados são


import sqlite3



# Para salvar a tabela no banco de dados



def salva_db (df, nome_db, nome_tabela):

    conn = sqlite3.connect(nome_db)

    df.to_sql(nome_tabela, conn, if_exists='replace', index=False)

    conn.close()

    return None
```
