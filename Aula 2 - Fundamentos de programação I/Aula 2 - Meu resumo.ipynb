## Variáveis

```python
# Variáveis tipo primitivo

var_int = 2 # Armazena apenas números inteiros
Var_float = 2.0 # Armazena apenas número decimais
var_str = "texto" # Armazena apenas texto
var_bool = True # Armazena apenas dois valores possíveis: "True" ou "False"

print (var_int)
print (Var_float)
print (var_str)
print (var_bool)
```

```python
# Para (for) cada letra no (in) nome, imprima (print) a letra

nome = 'Joao'
for letra in nome:
    print (letra)
```

```python
# Para (for) cada letra no (in) nome (variavel), imprima (print) a letra, porém (if) se ela for igual (==) A não imprima (pass)

nome = 'Joao'
for letra in nome:
  if letra == 'a':
    pass
  else:
    print (letra)
```


### f-strings

As f-strings (format strings) são uma característica introduzida no Python 3.6 que permite incorporar expressões e variáveis diretamente em strings formatadas de uma maneira mais legível e conveniente. Elas são criadas prefixando uma string com o caractere 'f' ou 'F' e, em seguida, usando chaves {} para incorporar expressões e variáveis dentro da string.


```python
nome = "João"
idade = 30

# Usando uma f-string para formatar a saída
mensagem = f"Olá, meu nome é {nome} e eu tenho {idade} anos."

print(mensagem)
```

**Operador AND**

O operador and em Python é um operador lógico usado para combinar duas ou mais expressões condicionais e retornar True apenas se todas as expressões forem avaliadas como verdadeiras. Se qualquer uma das expressões for avaliada como falsa, o operador and retornará False.

Aqui está a sintaxe básica do operador and:

expressão1 and expressão2

* Se expressão1 for avaliada como False, o resultado da operação será False, independentemente do valor de expressão2.
* Se expressão1 for avaliada como True, o resultado da operação será o valor de expressão2.

Aqui estão alguns exemplos:

```python
# Exemplo Aula  0 é menor(<) que X ?  e(AND) X é menor(<) que 100 ?

x = 10
print(0 < x and x < 100) #O resultado será verdadeiro pois 0 é menor que 10 e 10 é menor que 100


v = 20
print(0 < v and v > 100)
# Oresultado é falso, o primeiro parametro é verdadeiro pois 0 é menor que 20, porém o segundo é falso pois 20 não é maior que 100, então no final o resultado será falso,
#pois uma das condições é falsa. O resultado só aparece como True se todas as condições forem verdadeiras


# Exemplo 1
x = 5
y = 10
resultado = (x > 3) and (y < 15)
print(resultado)  # Isso imprimirá True porque ambas as expressões são verdadeiras.

# Exemplo 2
a = True
b = False
c = True
resultado = a and b and c
print(resultado)  # Isso imprimirá False porque b é falsa.

# Exemplo 3
valor = 42
resultado = (valor == 42) and (valor > 30)
print(resultado)  # Isso imprimirá True porque ambas as expressões são verdadeiras.
```

O operador and é usado em lógica condicional para exigir que todas as condições especificadas sejam verdadeiras antes que uma determinada ação seja tomada. É uma parte importante do controle de fluxo em Python e em muitas outras linguagens de programação.



---



**Operador OR**

O operador or em Python é um operador lógico usado para combinar duas ou mais expressões condicionais e retornar True se pelo menos uma das expressões for avaliada como verdadeira. Se todas as expressões forem avaliadas como falsas, o operador or retornará False.

Aqui está a sintaxe básica do operador or:

expressão1 or expressão2

* Se expressão1 for avaliada como True, o resultado da operação será True, independentemente do valor de expressão2.
* Se expressão1 for avaliada como False, o resultado da operação será o valor de expressão2.

Aqui estão alguns exemplos:

```python
# Exemplo 1
x = 5
y = 10
resultado = (x > 3) or (y < 5)
print(resultado)  # Isso imprimirá True porque pelo menos uma das expressões é verdadeira.

# Exemplo 2
a = False
b = True
c = False
resultado = a or b or c
print(resultado)  # Isso imprimirá True porque b é verdadeira.

# Exemplo 3
valor = 42
resultado = (valor == 42) or (valor > 100)
print(resultado)  # Isso imprimirá True porque a primeira expressão é verdadeira.
```

O operador or é usado em lógica condicional para tomar decisões com base em várias condições e é uma parte importante do controle de fluxo em Python e em muitas outras linguagens de programação.

**Operador NOT**

O operador not em Python é um operador lógico usado para negar ou inverter o valor de uma expressão condicional. Ele retorna True se a expressão for avaliada como falsa e False se a expressão for avaliada como verdadeira.

Aqui está a sintaxe básica do operador not:

not expressão

* Se expressão for avaliada como False, o resultado da operação not será True.
* Se expressão for avaliada como True, o resultado da operação not será False.

Aqui estão alguns exemplos:

```python
# Exemplo aula, o Operador not inverte o valor do resultado , exempo o resultado seria verdadeiro, inserindo o NOT ele seria convertido para false

a = 20

exemplo_aula = not (0 < a)  # Zero é menor que 20 - o resultado seria veradeiro, porém incluindo o not o resultado saira como false, pois ele vai inverter o valor do resultado
print(exemplo_aula)


# Exemplo 1
x = 5
resultado = not (x > 10)
print(resultado)  # Isso imprimirá True porque (x > 10) é falso.

# Exemplo 2
ativo = True
resultado = not ativo
print(resultado)  # Isso imprimirá False porque ativo é verdadeira.

# Exemplo 3
idade = 18
pode_votar = idade >= 18
nao_pode_votar = not pode_votar
print(nao_pode_votar)  # Isso imprimirá False porque idade >= 18 é verdadeira.
```

O operador not é usado para inverter o valor lógico de uma expressão, o que pode ser útil em condições condicionais, estruturas de controle de fluxo e lógica de programação em geral.
